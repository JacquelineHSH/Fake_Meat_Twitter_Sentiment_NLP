```python
# Read in twitter data just downloaded
artificial_meat_tweets = pd.read_csv('artificial meat.csv')
cultured_meat_tweets = pd.read_csv('cultured meat.csv')
beyond_meat_tweets = pd.read_csv('beyond meat.csv')
impossible_meat_tweets = pd.read_csv('impossible meat.csv')
```

```python
wanted_columns =  ['id','conversation_id','date','time', 'user_id', 'username', 'name','tweet'] 
```

```python
artificial_meat_tweets = artificial_meat_tweets[wanted_columns]
cultured_meat_tweets = cultured_meat_tweets[wanted_columns]
beyond_meat_tweets = beyond_meat_tweets[wanted_columns]
impossible_meat_tweets = impossible_meat_tweets[wanted_columns]

artificial_meat_tweets['search_key_word'] = 'artificial meat'
cultured_meat_tweets['search_key_word'] = 'cultured meat'
beyond_meat_tweets['search_key_word'] = 'beyond meat'
impossible_meat_tweets['search_key_word'] = 'impossible meat'
```

```python
fake_meat_tweets = pd.concat([artificial_meat_tweets,cultured_meat_tweets,beyond_meat_tweets,impossible_meat_tweets])
```

```python
fake_meat_tweets = fake_meat_tweets.drop_duplicates(subset=['id', 'conversation_id']) # Deduplicate
```

```python
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string
from nltk.tokenize import RegexpTokenizer, WordPunctTokenizer

# Define functions to do text preprocessing
tokenizer = RegexpTokenizer(r'\w+')

def remove_punctuation(text):
    no_punct = ''.join([c for c in text if c not in string.punctuation])
    return no_punct

stop_words = stopwords.words('english')

def remove_stopwords(text):
    words = [w for w in text if w not in stop_words]
    return words

lemmatizer = WordNetLemmatizer()

def word_lemmatizer(text):
    lem_text = [lemmatizer.lemmatize(i) for i in text]
    return lem_text

def str_to_lst(text):
    lst = text.lower().split()
    return lst

def lst_to_str(lst):
    string = ' '.join(lst)
    return string

def clean_tweets(tweet): # This is function from Jacqueline
    user_removed = re.sub(r'@[A-Za-z0-9]+','',tweet)
    link_removed = re.sub('https?://[A-Za-z0-9./]+','',user_removed)
    number_removed = re.sub('[^a-zA-Z]', ' ', link_removed)
    lower_case_tweet= number_removed.lower()
    tok = WordPunctTokenizer()
    words = tok.tokenize(lower_case_tweet)
    clean_tweet = (' '.join(words)).strip()
    return clean_tweet
#by row
```

```python
# Process the extracted dataframe's FormattedBody part
fake_meat_tweets['tweet'] = fake_meat_tweets['tweet'].astype(str)

fake_meat_tweets['tweet'] = fake_meat_tweets['tweet'].apply(lambda x: tokenizer.tokenize(x.lower()))

fake_meat_tweets['tweet'] = fake_meat_tweets['tweet'].apply(lambda x: remove_stopwords(x))

fake_meat_tweets['tweet'] = fake_meat_tweets['tweet'].apply(lambda x: word_lemmatizer(x))

fake_meat_tweets['tweet'] = fake_meat_tweets['tweet'].apply(lambda x: lst_to_str(x))

fake_meat_tweets['tweet'] = fake_meat_tweets['tweet'].apply(lambda x: clean_tweets(x))
```
